```python
from sarvamai import SarvamAI
from dotenv import load_dotenv
import os
import requests
from gtts import gTTS
import speech_recognition as sr
import soundfile as sf
from langdetect import detect
import base64
import playsound

load_dotenv()
headers={
    "api-subscription-key": os.getenv("SARVAM_API_KEY")
}
```

```python
#1)universal client object with the sarvamAI class with its api key 
#its constructor 
client=SarvamAI(api_subscription_key=os.getenv("SARVAM_API_KEY"))
```

```python

#1)text translation from one indic language to another using sarvamAI

def translate_text(text:str,source_lang:str,target_lang:str)->str:
    """
    Function to translate text from source language to target language
    using sarvamAI's translation API.
    
    Args:
    text (str): The text to be translated.
    source_lang (str): The source language code (e.g., 'hi' for Hindi).
    target_lang (str): The target language code (e.g., 'ta' for Tamil).
    
    Returns:
    str: Translated text.
    """
    return client.text.translate(
        input=text,
        source_language_code=source_lang,   #Input should be 'auto', 'en-IN', 'hi-IN', 'bn-IN', 'gu-IN', 'kn-IN', 'ml-IN', 'mr-IN', 'od-IN', 'pa-IN', 'ta-IN' or 'te-IN'
        target_language_code=target_lang,   #Input should be 'auto', 'en-IN', 'hi-IN', 'bn-IN', 'gu-IN', 'kn-IN', 'ml-IN', 'mr-IN', 'od-IN', 'pa-IN', 'ta-IN' or 'te-IN'
        speaker_gender="Female"
    )

## Example usage:
print(translate_text(
    text="नमस्ते, आप कैसे हैं?",
    source_lang="auto",  # Hindi
    target_lang="ta-IN"   # Tamil
).translated_text)
```

```python
#2) speech to text translation 
def speech_to_text_translation()->None:
    '''
    function to translate an input via microphone to text

    args:
    None

    returns:
    None
    '''
    indic_languages = {
        '1': ('Hindi', 'hi-IN'),
        '2': ('Tamil', 'ta-IN'),
        '3': ('Telugu', 'te-IN'),
        '4': ('Kannada', 'kn-IN'),
        '5': ('Malayalam', 'ml-IN'),
        '6': ('Marathi', 'mr-IN'),
        '7': ('Gujarati', 'gu-IN'),
        '8': ('Punjabi', 'pa-IN'),
        '9': ('Bengali', 'bn-IN'),
        '10': ('Odia', 'od-IN'),
        '11': ('Assamese', 'as-IN'),
        '12': ('Urdu', 'ur-IN')
    }
    # 1. Capture speech
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Please speak now...")
        audio = recognizer.listen(source)
        print("Processing audio...")

    try:
        spoken_text = recognizer.recognize_google(audio)
        print(f"\nYou said: {spoken_text}")
    except Exception as e:
        print(f"Could not recognize speech: {e}")
        return

    # 2. Detect language
    try:
        detected_lang = detect(spoken_text)
        print(f"Detected language code: {detected_lang}")
    except Exception as e:
        print(f"Could not detect language: {e}")
        detected_lang = 'en'

    # 3. Show target language menu
    print("\nAvailable Target Languages:")
    for key, (name, code) in indic_languages.items():
        print(f"{key}: {name} ({code})")
    choice = input("Select the target language by entering the corresponding number: ").strip()
    target_lang_code = indic_languages.get(choice, ('English', 'en-IN'))[1]  # Default to English 

    # 4. Compose source language code for SarvamAI
    source_lang_code = detected_lang + '-IN' if detected_lang != 'en' else 'en-IN'

    # 5. Translate
    translated_text = translate_text(spoken_text, source_lang_code, target_lang_code)

    # 6. Display results
    print(f"\nOriginal Text: {spoken_text}")
    print(f"Detected Language: {source_lang_code}")
    print(f"Target Language: {target_lang_code}")
    print(f"Translated Text: {translated_text.translated_text}")

# Call the function
speech_to_text_translation()
```

```python
#3) text to speech translation 
def text_to_speech(text: str, lang: str, output_file: str = "output.mp3") -> None:
    """
    Function to convert text to speech and play the audio using gTTS.

    Args:
        text (str): The text to be converted to speech.
        lang (str): The language code (e.g., 'nl' for Dutch).
        output_file (str): The filename for the output MP3 file.

    Returns:
        None
    """
    tts = gTTS(text=text, lang=lang)
    tts.save(output_file)
    os.system(f"start {output_file}")  # For Windows. Use 'afplay' for Mac or 'mpg123' for Linux.

text_to_speech("hello i am rex joson","en")
```

```python

from pydub import AudioSegment
from pydub.playback import play

def speech_to_speech_translation() -> None:
    '''
    Function to translate an input via microphone to speech in another language and save as WAV.

    Args:
        None
    Returns:
        None
    '''
    indic_languages = {
        '1': ('Hindi', 'hi-IN'),
        '2': ('Tamil', 'ta-IN'),
        '3': ('Telugu', 'te-IN'),
        '4': ('Kannada', 'kn-IN'),
        '5': ('Malayalam', 'ml-IN'),
        '6': ('Marathi', 'mr-IN'),
        '7': ('Gujarati', 'gu-IN'),
        '8': ('Punjabi', 'pa-IN'),
        '9': ('Bengali', 'bn-IN'),
        '10': ('Odia', 'od-IN'),
        '11': ('Assamese', 'as-IN'),
        '12': ('Urdu', 'ur-IN')
    }

    # 1. Capture speech
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Please speak now...")
        audio = recognizer.listen(source)
        print("Processing audio...")

    try:
        spoken_text = recognizer.recognize_google(audio)
        print(f"\nYou said: {spoken_text}")
    except Exception as e:
        print(f"Could not recognize speech: {e}")
        return

    # 2. Detect language
    try:
        detected_lang = detect(spoken_text)
        print(f"Detected language code: {detected_lang}")
    except Exception as e:
        print(f"Could not detect language: {e}")
        detected_lang = 'en'

    # 3. Show target language menu
    print("\nAvailable Target Languages:")
    for key, (name, code) in indic_languages.items():
        print(f"{key}: {name} ({code})")
    choice = input("Select the target language by entering the corresponding number: ").strip()
    target_lang_code = indic_languages.get(choice, ('English', 'en-IN'))[1]

    # 4. Compose source language code for SarvamAI
    source_lang_code = detected_lang + '-IN' if detected_lang != 'en' else 'en-IN'

    # 5. Translate
    translation_response = translate_text(spoken_text, source_lang_code, target_lang_code)
    
    # 6. Extract the actual text from the response object
    translated_text = translation_response.translated_text

    # 7. Text-to-Speech and convert to WAV
    tts_lang = target_lang_code.split('-')[0]
    tts = gTTS(text=translated_text, lang=tts_lang)
    
    # Save as MP3 first (gTTS limitation)
    tts.save("temp_output.mp3")
    
    # Convert MP3 to WAV
    audio_segment = AudioSegment.from_mp3("temp_output.mp3")
    audio_segment.export("translated_output.wav", format="wav")
    
    # Clean up temporary MP3 file
    os.remove("temp_output.mp3")
    
    print(f"\nWAV file saved as 'translated_output.wav'")
    print(f"Playing translated speech in {target_lang_code}...")
    
    # Play the WAV file
    play(audio_segment)

    # 8. Display results
    print(f"\nOriginal Text: {spoken_text}")
    print(f"Detected Language: {source_lang_code}")
    print(f"Target Language: {target_lang_code}")
    print(f"Translated Text: {translated_text}")
    print(f"Output saved as: translated_output.wav")

# Call the function
try:
    speech_to_speech_translation()
except:
    pass    
```
